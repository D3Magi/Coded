```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import warnings
warnings.filterwarnings ('ignore')
```

```python
AW_data= pd.read_csv('C:/Users/user/Desktop/Assignment Weather AUSTRALIA/weatherAUS.csv')
AW_test_data= pd.read_csv('C:/Users/user/Desktop/Assignment Weather AUSTRALIA/test.csv')
```

```python
#using the label encoder
le=LabelEncoder()
```

```python
cat_col = [i for i in AW_data.columns if AW_data[i].dtype == "object"]
```

```python
#converting the categorical columns to numerical data using label encoder
for col in cat_col:
    AW_data[col] = le.fit_transform(AW_data[col])
```

```python
print(AW_data)
```

```python
AW_data_filled_mean = AW_data.fillna(AW_data.mean())
```

```python
print(AW_data_filled_mean)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X= AW_data_filled_mean.drop(columns=['RainTomorrow'])
y= AW_data_filled_mean['RainTomorrow']
```

```python
#use train test split to split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
#Initialize the KNeighbours
model_KNN = KNeighborsClassifier()
model_KNN.fit(X_train, y_train)
```

```python
#make predictions using the KNN classifier
y_predict_KNN= model_KNN.predict(X_test)
Accuracy_KNN= accuracy_score(y_test, y_predict_KNN)
print("Accuracy Score:")
print(Accuracy_KNN)
```

```python
# Initialize the Decision Tree Classifier
model_DT = DecisionTreeClassifier()
model_DT.fit(X_train, y_train) #the train data for x and y are used to train

#make predictions
y_predict_DT = model_DT.predict(X_test)

Accuracy_DT = accuracy_score(y_test, y_predict_DT)
print("Accuracy Score:")
print(Accuracy_DT)
```

```python
# Initialize and train RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)
```

```python
# Predict on the test set
y_predict_rf = model_rf.predict(X_test)

Accuracy_RT= accuracy_score(y_test, y_predict_rf)
print("Accuracy score")
print(Accuracy_RT)
```

From the various models used for the prediction of the weather in Australia, it can be seen that the Random forest is preferable, followed by the KNN Classifier. These models gave a 84% and 81% accuracy score respoectively.
